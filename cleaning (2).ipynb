```python
df = pd.read_csv("deceptive-opinion.csv")
print(f"Shape: {df.shape}")
df.head()
```

```python
df.isnull().sum()
```

```python
def clean_text(text):
    text = text.replace(',', '')
    text = text.lower()
    words = text.split()
    words = [w for w in words if w not in stop_words]
    return ' '.join(words)

df['text'] = df['text'].apply(clean_text)
print("Text cleaning complete.")
```
